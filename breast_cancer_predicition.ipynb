```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_datasets = sklearn.datasets.load_breast_cancer()
```

```python
x = data_frame.drop(columns = 'lable',axis = 1)
y = data_frame['lable']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_sd = scaler.fit_transform(x_train)
x_test_sd = scaler.fit_transform(x_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(100,input_shape = (30,),activation = 'relu'),
    keras.layers.Dense(100 ,activation = 'relu'),
    keras.layers.Dense(100,activation = 'softmax')
])
```

```python
model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
history = model.fit(x_train_sd,y_train,validation_split = 0.1,epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(x_test_sd, y_test)
print(accuracy)
```
